```python
import numpy as np
import pylab as plt

epochs = 36511

dropout_3 = np.load('3 layers with dropout.npy') 
no_dropout_3 = np.load('3 layers without dropout.npy')
dropout_4 = np.load('4 layers with dropout.npy') 
no_dropout_4 = np.load('4 layers without dropout.npy')
dropout_5 = np.load('5 layers with dropout.npy') 
no_dropout_5 = np.load('5 layers without dropout.npy')

plt.figure(figsize=(15,10))
plt.title('errors vs. epochs')
plt.xlabel('epochs')
plt.ylabel('errors')
plt.xlim(0,36511)
plt.ylim(0,0.08)
plt.plot(range(epochs), dropout_3,'b',label='3-layer NN with dropout')
plt.plot(range(epochs), no_dropout_3,'g',label='3-layer NN without dropout')
plt.plot(range(epochs), dropout_4,'r',label='4-layer NN with dropout')
plt.plot(range(epochs), no_dropout_4,'c',label='4-layer NN without dropout')
plt.plot(range(epochs), dropout_5,'m',label='5-layer NN with dropout')
plt.plot(range(epochs), no_dropout_5,'y',label='5-layer NN without dropout')
plt.legend(loc='best')
plt.show()  
```
