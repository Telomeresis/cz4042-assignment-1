```python
import tensorflow as tf
import numpy as np
import pylab as plt
import pandas as pd
import sys

np.set_printoptions(threshold=sys.maxsize)
tf.compat.v1.logging.set_verbosity(tf.compat.v1.logging.ERROR)
```

```python
# Set compatibility mode
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

# Set pandas display option
pd.options.display.float_format = "{:,.4f}".format 

# RTX ON
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        # Currently, memory growth needs to be the same across GPUs
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        logical_gpus = tf.config.experimental.list_logical_devices('GPU')
        print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
    except RuntimeError as e:
        # Memory growth must be set before GPUs have been initialized
        print(e)
```

```python
# scale data
def scale(X):
    return (X-np.mean(X,axis=0)) / np.std(X,axis=0)
```

```python
NUM_FEATURES = 7

hidden_neurons = 10
learning_rate = 10**-3
beta = 10**-3
batch_size = 8

epochs = 36511
seed = 42
np.random.seed(seed)
```

```python
# Read data
data_input = np.genfromtxt('admission_predict.csv', delimiter= ',')
data_X, data_Y = data_input[1:,1:8], data_input[1:,-1]
# Scale X inputs
dataX = scale(data_X)
dataY = data_Y.reshape(data_Y.shape[0],1)

# Define the size of training set and data set
dataset_size = len(dataX)
train_size = int(0.7 * dataset_size)
test_size = dataset_size - train_size

# Shuffle the data before splitting to train test set
idx = np.arange(dataset_size)
np.random.shuffle(idx)
dataX = dataX[idx]
dataY = dataY[idx]

# Split the training and testing set
trainX = dataX[:train_size]
trainY = dataY[:train_size]

testX = dataX[train_size:]
testY = dataY[train_size:]
```

## Model using all 7 input features
---

```python
NUM_FEATURES = 7

# Create the model
x = tf.placeholder(tf.float32, [None, NUM_FEATURES])
y_ = tf.placeholder(tf.float32, [None, 1])

# Build the graph for the neural network

w1 = tf.Variable(tf.truncated_normal([NUM_FEATURES, hidden_neurons], stddev=1.0 / np.sqrt(NUM_FEATURES), dtype=tf.float32), name='weights')
b1 = tf.Variable(tf.zeros([hidden_neurons]), dtype=tf.float32, name='biases')
h1 = tf.nn.relu(tf.matmul(x, w1) + b1) #Hidden layer with relu activation

w2 = tf.Variable(tf.truncated_normal([hidden_neurons, 1], stddev=1.0 / np.sqrt(hidden_neurons), dtype=tf.float32), name='weights')
b2 = tf.Variable(tf.zeros([1]),dtype=tf.float32, name='biases')
y = tf.matmul(h1, w2) + b2  

# Regularization
regularization = tf.nn.l2_loss(w1) + tf.nn.l2_loss(w2)

# Regularized loss
loss = tf.reduce_mean(tf.square(y_ - y) + beta*regularization)

# Define the optimiser
optimizer = tf.train.GradientDescentOptimizer(learning_rate)

# Create a variable to track the global step
global_step = tf.Variable(0, name='global_step', trainable=False)

# Use the optimizer to apply the gradients that minimize the loss
# (and also increment the global step counter) as a single training step.
train_op = optimizer.minimize(loss, global_step=global_step)
```

```python
N = train_size
idx_train = np.arange(N)

# Run the session
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

# Define an array to store the test errors during training
test_error_7 = np.zeros(epochs)

for i in range(epochs):
    np.random.shuffle(idx_train) 
    trainX = trainX[idx_train]
    trainY = trainY[idx_train]
  
    for start, end in zip(range(0, N, batch_size), range(batch_size, N, batch_size)):
        sess.run(train_op, {x:trainX[start:end],y_: trainY[start:end]})
    
    t_err = sess.run(loss, {x: testX, y_: testY}) 
    test_error_7[i] = t_err

    if i == 0 or (i+1)%5000 == 0 or i==epochs-1:
        print('Iteration {}: test error {}'.format(i+1, test_error_7[i]))
```

## Model using 6 input features

```python
NUM_FEATURES = 6

# Create the model
x = tf.placeholder(tf.float32, [None, NUM_FEATURES])
y_ = tf.placeholder(tf.float32, [None, 1])

# Build the graph for the deep neural network
# Hidden layer with relu activation
w1 = tf.Variable(tf.truncated_normal([NUM_FEATURES, hidden_neurons], stddev=1.0 / np.sqrt(NUM_FEATURES), dtype=tf.float32), name='weights')
b1 = tf.Variable(tf.zeros([hidden_neurons]), dtype=tf.float32, name='biases')
h1 = tf.nn.relu(tf.matmul(x, w1) + b1) 

w2 = tf.Variable(tf.truncated_normal([hidden_neurons, 1], stddev=1.0 / np.sqrt(hidden_neurons), dtype=tf.float32), name='weights')
b2 = tf.Variable(tf.zeros([1]),dtype=tf.float32, name='biases')
y = tf.matmul(h1, w2) + b2  

# Regularization
regularization = tf.nn.l2_loss(w1) + tf.nn.l2_loss(w2)

# Regularized loss
loss = tf.reduce_mean(tf.square(y_ - y) + beta*regularization)

# Define the optimiser
optimizer = tf.train.GradientDescentOptimizer(learning_rate)

# Create a variable to track the global step
global_step = tf.Variable(0, name='global_step', trainable=False)

# Use the optimizer to apply the gradients that minimize the loss
# (and also increment the global step counter) as a single training step.
train_op = optimizer.minimize(loss, global_step=global_step)
```

```python
N = train_size
idx_train = np.arange(N)

# Define an array to store the test errors during training
test_error_6 = np.zeros((NUM_FEATURES+1,epochs))

# Remove one feature each iteration
for feature in range(NUM_FEATURES+1):
    print('Removed input feature', feature)

    # Remove one input feature from the training and test data
    trainX_6 = np.delete(trainX, feature, 1)
    testX_6 = np.delete(testX, feature, 1)

    # Run the session
    sess = tf.Session()
    init = tf.global_variables_initializer()
    sess.run(init)
  
    for i in range(epochs):
        np.random.shuffle(idx_train) 
        trainX_6 = trainX_6[idx_train]
        trainY = trainY[idx_train]

        for start, end in zip(range(0, N, batch_size), range(batch_size, N, batch_size)):
            sess.run(train_op, {x:trainX_6[start:end], y_:trainY[start:end]})

        t_err = sess.run(loss, {x: testX_6, y_: testY}) 
        test_error_6[feature][i] = t_err

        if i == 0 or (i+1)%5000 == 0 or i==epochs-1:
            print('Iteration {}: test error {}'.format(i+1, test_error_6[feature][i]))

    print('')
```

```python
# Plot learning curves
plt.figure(figsize=(20,10))
plt.title('test errors vs. epochs')
plt.plot(range(epochs), test_error_7, 'r', linestyle='--', label='all features')
plt.plot(range(epochs), test_error_6[0], 'b', label='remove GRE Score')
plt.plot(range(epochs), test_error_6[1], 'g', label='remove TOEFL Score')
plt.plot(range(epochs), test_error_6[2], 'c', label='remove University Rating')
plt.plot(range(epochs), test_error_6[3], 'r', label='remove SOP')
plt.plot(range(epochs), test_error_6[4], 'm', label='remove LOR')
plt.plot(range(epochs), test_error_6[5], 'y', label='remove CGPA')
plt.plot(range(epochs), test_error_6[6], 'k', label='remove Research')

plt.xlabel('epochs')
plt.ylabel('test errors')
plt.legend(loc='best')
plt.ylim((0,0.035))

plt.show()
```

```python
import time

N = train_size
idx_train = np.arange(N)

# Define an array to store the test errors during training
test_error_5 = np.zeros((NUM_FEATURES+1,epochs))

# Calculate time take for execution
start_t = time.time()

# Remove one feature each iteration
for feature in range(NUM_FEATURES+1):
    print('Removed input feature', feature)

    # Remove one input feature from the training and test data
    trainX_5 = np.delete(trainX_6, feature, 1)
    testX_5 = np.delete(testX_6, feature, 1)

    # Run the session
    sess = tf.Session()
    init = tf.global_variables_initializer()
    sess.run(init)
  
    for i in range(epochs):
        np.random.shuffle(idx_train) 
        trainX_5 = trainX_5[idx_train]
        trainY = trainY[idx_train]

        for start, end in zip(range(0, N, batch_size), range(batch_size, N, batch_size)):
            sess.run(train_op, {x:trainX_5[start:end], y_:trainY[start:end]})

        t_err = sess.run(loss, {x: testX_5, y_: testY}) 
        test_error_5[feature][i] = t_err

        if i == 0 or (i+1)%5000 == 0 or i==epochs-1:
            print('Iteration {}: test error {}'.format(i+1, test_error_5[feature][i]))

end_t = time.time()
```
